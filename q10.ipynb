```python
import math
```

```python
def CheckIfPrime(num):
    if num % 2 == 0:
        return False
    else:
        for j in range(1,int(math.floor(num/2))+1):
            n = 2*j-1
            if (num % n == 0) and (n != (1 or num)):
                return False
    return True
```

```python
def AllPrimesWithin(num):
    Primes = [2]
    current = 3
    while (current < num):
        if CheckIfPrime(current):
            Primes.append(current)
        current = current + 2
    return Primes
```

```python
sum(AllPrimesWithin(10))
```
