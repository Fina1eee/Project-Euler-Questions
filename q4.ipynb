```python
import numpy as np
import math
```

```python
def IsPalindromic(num):
    NumAsString = str(num)
    StringLength = len(NumAsString)
    for i in range(math.floor(StringLength/2)):
        if NumAsString[i] != NumAsString[StringLength - 1 - i]:
            return 0
    return 1
```

```python
Palin = []
for i in range(100,999):
    for j in range(100,999):
        num = i * j
        if (IsPalindromic(num) == 1):
            Palin.append(num)
print(max(Palin))

        
```
