```python
def SumOfSquares(num):
    sum = 0
    for i in range(1,num+1):
        sum = sum + i**2
    return sum
```

```python
def SquareOfSum(num):
    sum = 0
    for i in range(1,num+1):
        sum = sum + i
    sum = sum**2
    return sum
```

```python
print(SquareOfSum(100)-SumOfSquares(100))
```
