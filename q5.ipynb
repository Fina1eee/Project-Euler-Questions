```python
def isNotDivisableOneThru20(num):
    for i in range(1,21):
        if num % i != 0:
            return True #This condition will keep the while loop going
    print("{} is Divisible By All Numbers Between 1 and 20!".format(num))
    return False
```

To Avoid Testing Every Possible Number, I will be only testing multiples of 20. This will save a lot of processing power.
