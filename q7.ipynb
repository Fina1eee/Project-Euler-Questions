```python
import numpy as np
import math
```

```python
def CheckIfPrime(num):
    if num % 2 == 0:
        return False
    else:
        for j in range(1,int(math.floor(num/2))+1):
            n = 2*j-1
            if (num % n == 0) and (n != (1 or num)):
                return False
    return True
```

```python
def FindnthPrimeNumber(n):
    if n == 1:
        return 2
    if n == 2:
        return 3
    i = 2
    num = 3
    while (i < n):
        num = num + 2
        if CheckIfPrime(num):
            i = i + 1
    return num
            

```
