```python
import numpy as np
```

```python
def IsPrime(num):
    for i in range(2,num):
        if (num % i == 0):
            if i != num:
                return 0
    return 1
```

```python
def AllPrimesWithin(num):
    Primes = []
    for i in range(2,num):
        if IsPrime(i) == 1:
            Primes.append(i)
    return Primes
```

```python
AllPrimesWithin(10)
```

```python
num = 13195
Primes = AllPrimesWithin(num)
PrimeFactors = []
for i in range(len(Primes)):
    if num % Primes[i] == 0:
        PrimeFactors.append(Primes[i])
print(PrimeFactors)
print(max(PrimeFactors))
```
